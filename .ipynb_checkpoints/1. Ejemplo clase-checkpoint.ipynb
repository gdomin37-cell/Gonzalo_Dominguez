### String literals

El hecho de que el contenido de las variables de tipo `str` vaya entre comillas, ya sean simples o dobles, conlleva a que algunos caracteres deban ser tratados de forma especial. 

Aquí entran en juego los string literals. Algunos de los más utilizados se muestran en la siguiente tabla:

Código | Significado |
--- | --- |
`\\` | Backslash, `\`| 
`\'` | Comilla simple, `'`| 
`\"` | Comilla doble, `"`|  
`\n` | Salto de línea| 
`\t` | Tabulación horizontal|

```python
s3 = "Primer renglon \n Segundo renglon"
s3
```

```python
print(s3)
```

Las cadenas se comparan carácter por carácter según su valor Unicode (o ASCII). Esto significa que:

1) Se compara el primer carácter de cada string.

2) Si son iguales, pasa al siguiente.

3) La comparación es sensible a mayúsculas y minúsculas (case-sensitive).

 Las letras mayúsculas vienen antes que las minúsculas.

### Condicionales - IF

```python
edad = 25
```

```python
if edad == 25:
    print("Tiene 25 años")
```

```python
if (edad == 25){
    print("Tiene 25 años")
} else if (edad < 25) {
    print("No tiene 25 años")
}
```

```python
if edad == 25:
    print("Tiene 25 años")
else:
    print("No tiene 25 años")
```

```python
if edad == 25:
    print("Tiene 25 años")
elif edad < 25:
    print("Tiene menos de 25 años")
else:
    print("Tiene más de 25 años")
```

```python
edad = 16
altura = 140

if edad >= 18 and (altura >= 150 and altura < 200):
    print("Que pase")
elif edad >= 18 and altura < 150:
    print("Entre bajo su propio riesgo")
elif edad < 18 and altura >= 150:
    print("Entre con un mayor")
else:
    print("Andate a tu casa")
```

## Continue

```python
for i in range(30):
    #i=4
    if i % 3 == 0:
        continue
        print("ESTO NO SE VA A IMPRIMIR NUNCA")
print(i)
```

```python
variable_g = range(30)
for a in variable_g:
    if a % 3 == 0:
        continue # its a bypass and doesnt take any action
        print("Esto no se imprime")
    print(a)
```

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1,10):
    #i=4
    if i % 3 == 0:
        break
        
    print(i)
```
